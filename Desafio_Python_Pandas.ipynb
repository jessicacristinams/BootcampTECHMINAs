3.Qual a média de valores de empresa das cinco categorias mais valiosas?


```python
# Categorias mais valiosas

cat_valiosas = pd.DataFrame(FORBES.groupby('category').marketvalue.sum()).nlargest(5,"marketvalue")
cat_valiosas
```

1. Crie uma função que ao informar o sexo e a faixa etária de um suícidio é retornado as probabilidades do método utilizado.

```python
def suicide_func(sex,age):
  data2 = df_suicide.copy()
  data2 = data2[data2['sex'] == sex ]
  data2 = data2[data2['age.group'] == age ]
  data2['Freq'] = data2['Freq']/data2['Freq'].sum()
  data2 = data2.rename(columns={'method2':'Metodo utilizado','Freq':'Probabilidade'})
  data2 = data2.reset_index(drop=True)
 # print(data2['Probabilidade'].sum())            verificação da soma das probabilidades
  return(data2)
```

```python
#criação de função auxiliar que converte a idade a uma faixa etária
def age_to_agegroup(idade):
  if idade >=10 and idade <=20:
    age_group = '10-20'
  elif idade >=25 and idade <=35:
    age_group = '25-35'
  elif idade >=40 and idade <=50:
    age_group = '40-50'
  elif idade >=55 and idade <=65:
    age_group = '55-65'
  elif idade >=70 and idade <=90:
    age_group = '70-90'
  else: 
    age_group = 'fora do escopo do estudo'
    print(age_group)
  return(age_group)
```

```python
# função para definir o sexo digitado pelo usuário

def insert_sex():
  sex_number = 0
  while sex_number not in [1,2]:
    sex_number = int(input("Insira o sexo: 1 - feminino, 2 - masculino: "))
    if sex_number == 1:
      sex = 'female'
    elif sex_number == 2:
      sex = 'male'
    else:
      print('Opção inválida')
  return sex
```

```python
#função para definir a faixa etária desejada pelo usuário

def def_faixaetaria():
  faixa_etaria = 'fora do escopo do estudo'
  while(faixa_etaria == 'fora do escopo do estudo'):
    age = int(input("Insira a idade para retornar a faixa etária desejada: "))
    faixa_etaria = age_to_agegroup(age)
  return faixa_etaria
```
